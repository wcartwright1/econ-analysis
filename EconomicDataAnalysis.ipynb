```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.io as pio
import os.path
import datetime as dt
```

USER INPUT PARAMETERS

```python
# User input for the base location of the project
USER_BASE_LOCATION = r'C:\Users\billy\Documents\01_HGVoraExercise'

# User input for the relative location of the input data and where the results will be output
USER_INPUT_DATA_LOCATION = r'03_Input'
USER_OUTPUT_DATA_LOCATION = r'04_Output'

# User input for the name of the data sources used in the project
USER_INPUT_CSV_1 = r'MEDCPIM158SFRBCLE.csv' # Median Consumer Price Index https://fred.stlouisfed.org/series/MEDCPIM158SFRBCLE
USER_INPUT_CSV_2 = r'REAINTRATREARAT10Y.csv '# 10-Year Real Interest Rate https://fred.stlouisfed.org/series/REAINTRATREARAT10Y

# User input for data column names to allow transferability of code with new datasets
USER_INPUT_DATA_X_VARIABLE_1 = 'DATE'
USER_INPUT_DATA_Y_VARIABLE_1 = 'MEDCPIM158SFRBCLE'
USER_INPUT_DATA_Y_VARIABLE_DESC_1 = 'Median Consumer Price Index'
USER_INPUT_DATA_X_VARIABLE_2 = 'DATE'
USER_INPUT_DATA_Y_VARIABLE_2 = 'REAINTRATREARAT10Y'
USER_INPUT_DATA_Y_VARIABLE_DESC_2 = '10-Year Real Interest Rate'
```

DATA SOURCING - PATH DEFINITION

```python
# Define location of input and output data based on user input location
inputLocation = f'{USER_BASE_LOCATION}\\{USER_INPUT_DATA_LOCATION}'
outputLocation = f'{USER_BASE_LOCATION}\\{USER_OUTPUT_DATA_LOCATION}'
```

```python
# Define full input and output file paths
inputFilePath1 = f'{inputLocation}\\{USER_INPUT_CSV_1}'
inputFilePath2 = f'{inputLocation}\\{USER_INPUT_CSV_2}'
```

DATA SOURCING - INPUT AND PREPROCESSING

```python
# Source csv files as pandas dataframes
df_1input = pd.read_csv(inputFilePath1, sep=',')
df_2input = pd.read_csv(inputFilePath2, sep=',')
```

```python
# Define columns based on user input column names
df_1inputPreproc = df_1input.copy()
df_1inputPreproc = df_1inputPreproc.melt(id_vars=['DATE'], value_vars=USER_INPUT_DATA_Y_VARIABLE_1)
df_1inputPreproc['variable_desc'] = USER_INPUT_DATA_Y_VARIABLE_DESC_1

df_2inputPreproc = df_2input.copy()
df_2inputPreproc = df_2inputPreproc.melt(id_vars=['DATE'], value_vars=USER_INPUT_DATA_Y_VARIABLE_2)
df_2inputPreproc['variable_desc'] = USER_INPUT_DATA_Y_VARIABLE_DESC_2
```

```python
# Stack dataframes for plotting
df_inputStack = pd.concat([df_1inputPreproc, df_2inputPreproc])
```

```python
# Create simple line chart plotting the value of each economic indicator over time
fig_simpleLineChart = px.line(df_inputStack, 
                              x='DATE', 
                              y='value', 
                              color='variable_desc', 
                              title=f'{USER_INPUT_DATA_Y_VARIABLE_DESC_1} & {USER_INPUT_DATA_Y_VARIABLE_DESC_2}'
                              )
```

```python
# Display simple line chart using plotly in Jupyter Notebook
fig_simpleLineChart.show()
```

```python
pio.write_image(fig_simpleLineChart, f'{outputLocation}\\simpleLineChart.pdf')
```
